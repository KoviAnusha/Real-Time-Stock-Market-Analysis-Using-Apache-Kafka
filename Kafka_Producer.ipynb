```python
import pandas as pd
from kafka import KafkaProducer
from time import sleep
import json
from json import dumps
```

```python
producer = KafkaProducer(bootstrap_servers=['18.220.138.198:9092'], value_serializer=lambda x: dumps(x).encode('utf-8'))
```

```python
producer.send('demo_testing_topic',value="{'hellowwwwww':'world'}")
```
